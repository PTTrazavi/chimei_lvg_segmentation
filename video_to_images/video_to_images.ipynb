## Import libraries and set parameters

```python
import cv2
import matplotlib.pyplot as plt
import os

# folder_name = '../data/2020_1223_MASK_LV/20458/frames/'
folder_name = '../data/20210120/0747 basal hypo/0747_frames/'
if not os.path.exists(folder_name):
    os.mkdir(folder_name)
# video_file = '../data/2020_1223_MASK_LV/20458/20458.avi'
video_file = '../data/20210120/0747 basal hypo/0747.avi'

crop_height = 712 # 892
crop_width = 712 # 892

target_height = 512
target_width = 512
```

## Check the location of the wave

```python
# Reading in video
vidcap = cv2.VideoCapture(video_file)
success = True
count = 0

while success:
    # read a frame
    success, image = vidcap.read()
    print("image size:", image.shape)
    print('Read a new frame : ', success)
    count += 1
    
    if count == 1:
        plt.imshow(image)
        break
```

## Crop test

```python
# Reading in video
vidcap = cv2.VideoCapture(video_file)
success = True
count = 0

while success:
    # read a frame
    success, image = vidcap.read()
    print("image size:", image.shape)
    print('Read a new frame : ', success)
    # crop the image
    image = image[:crop_height, :crop_width]
    print('cropped image size : ', image.shape)
    # resize the image
    if image.shape[0] != target_height or image.shape[1] != target_width:
        image = cv2.resize(image, (target_height, target_width))
        print("resized image size:", image.shape)
    # add 1 to counter
    count += 1
    
    if count == 1:
        plt.imshow(image)
        break
```

## Read and save the frames

```python
# Reading in video
vidcap = cv2.VideoCapture(video_file)
success = True
count = 0

while success:
    # read a frame
    success, image = vidcap.read()
    print("image size:", image.shape)
    print('Read a new frame : ', success)
    # crop the image
    image = image[:crop_height, :crop_width]
    # resize the image
    if image.shape[0] != target_height or image.shape[1] != target_width:
        image = cv2.resize(image, (target_height, target_width))
    # save the frame 
    save_name = '0'*(5-len(str(count))) +  str(count) + '.jpg'
    cv2.imwrite(folder_name + save_name , image)
    # add 1 to counter
    count += 1
    print('Read a new frame : ', save_name)
```
